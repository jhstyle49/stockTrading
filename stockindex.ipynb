# Module

```python
import yfinance as yf
import pandas as pd
import numpy as np
```

# Month begin stockprice by ticker 

```python
ticker_usd = ['^GSPC', 'NDX', 'SCHD', 'O', 'SBUX', 'GOOG', 'MSFT']
result=yf.download(ticker_usd,start = '2020-01-01')
result.columns = ['_'.join(c) for c in result.columns]
result = result.reset_index()
result['Date'] = result['Date'].dt.strftime('%Y-%m-%d')
result['Date'] = pd.to_datetime(result['Date'])

remain_cols = []
for t in ticker_usd:
    remain_cols.append(f"Close_{t}")

result = result[['Date']+remain_cols]
result = result.dropna()
result['month_start'] = result['Date'] + pd.offsets.MonthEnd(0) + pd.Timedelta(days=1)
result = result.drop_duplicates(subset='month_start', keep='last')
result.rename(columns=dict(zip([c for c in result.columns if 'Close_' in c], ticker_usd)), inplace=True)
result.rename(columns={'Date':'Date_USD'}, inplace=True)
df_usd = result.copy()
df_usd
```

```python
result[result['Date_USD'].dt.month.isin([8,9])].to_excel("9month_stock.xlsx")
```

```python
ticker_ko = ['^KS11', 'USDKRW=X']
result=yf.download(ticker_ko,start = '2023-11-01')
result
result.columns = ['_'.join(c) for c in result.columns]
result = result.reset_index()
result['Date'] = result['Date'].dt.strftime('%Y-%m-%d')
result['Date'] = pd.to_datetime(result['Date'])

remain_cols = []
for t in ticker_ko:
    remain_cols.append(f"Close_{t}")

result = result[['Date']+remain_cols]
result = result.dropna()
result['month_start'] = result['Date'] + pd.offsets.MonthBegin(0)
result = result.drop_duplicates(subset='month_start', keep='last')
result.rename(columns=dict(zip([c for c in result.columns if 'Close_' in c], ticker_ko)), inplace=True)
result.rename(columns={'Date':'Date_ko'}, inplace=True)
df_korea = result.copy()
df_korea
```

```python
df = pd.merge(df_korea, df_usd, on='month_start')
```

```python
ticker = ticker_ko + ticker_usd
display(ticker)
stock = ['KOSPI', 'USDKRW', 'S&P500', 'NASDAQ100', 'SCHD', 'O', 'SBUX']
df = df[['Date_ko', 'Date_USD', 'month_start']+ticker]
df.rename(columns=dict(zip(ticker, stock)), inplace=True)
df = df.round(1)
```

```python
df.to_excel("241201.xlsx")
```

# MDD, YTD

```python
ticker_usd = ['^KS11', '^GSPC', 'NDX', 'SCHD', 'O', 'SBUX', 'GOOG', 'MSFT']
```

```python
result=yf.download(ticker_usd,start = '2022-12-01')
result.columns = ['_'.join(c) for c in result.columns]
result = result.reset_index()
result['Date'] = result['Date'].dt.strftime('%Y-%m-%d')
result['Date'] = pd.to_datetime(result['Date'])

remain_cols = []
for t in ticker_usd:
    remain_cols.append(f"Close_{t}")

result = result[['Date']+remain_cols]
# result = result.dropna()
for c in result.columns:
    result[c] = result[c].fillna(method='ffill')

result.rename(columns=dict(zip([c for c in result.columns if 'Close_' in c], ticker_usd)), inplace=True)
result.rename(columns={"^KS11": "KOSPI",'^GSPC':'S&P500', 'NDX':'NASDAQ100'}, inplace=True)
```

```python
df_mdd = pd.concat([result[result['Date']=='2023-12-29'].iloc[:, 1:].T,  pd.DataFrame(result.iloc[:,1:].max()), pd.DataFrame(result.iloc[-1,1:])], axis=1)
df_mdd.columns = ['2024Begin', 'max_price', 'price']
df_mdd['mdd'] = df_mdd.apply(lambda x: np.round((x['price']/x['max_price']-1)*100, 1), axis=1)
df_mdd['YTD'] = df_mdd.apply(lambda x: (x['price']/x['2024Begin']-1)*100, axis=1)

for c in df_mdd.columns:
    df_mdd[c] = pd.to_numeric(df_mdd[c])
    df_mdd[c] = df_mdd[c].round(1)
df_mdd.reset_index().rename(columns={'index':'stock'})
```

```python
with pd.ExcelWriter("marketprice.xlsx", engine='xlsxwriter') as writer:
    df.to_excel(writer, sheet_name='price', index=False)
    df_mdd.to_excel(writer, sheet_name='mdd_ytd_price', index=False)
```
